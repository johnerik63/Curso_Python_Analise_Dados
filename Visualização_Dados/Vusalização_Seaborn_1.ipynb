```python
import numpy as np
import pandas as pd
import seaborn as sns

# definir tema 
sns.set_theme(style='darkgrid')
```

```python
# definindo base de dados 
Base_Dados = sns.load_dataset('tips')

#Verificar 

Base_Dados.head()
```

```python
#renomeando as colunas

Base_Dados.rename(columns= {
    'total_bill':'Total Conta',
    'tip':'Gorjeta',
    'sex':'Sexo',
    'smoker':'Fumante',
    'day':'Dia Semana',
    'time':'Peíodo',
    'size':'Pessoas na Mesa'
})
```

```python
#gráfico relplot

sns.relplot( x= 'Total Conta', y = 'Gorjeta', data=Base_Dados);
# insigth = Podemos notar que quanto maior o valor da conta, maior a gorjeta recebida pelo garçom
```

```python
# Passando outro parâmetro gráfico como classe no gráfico

sns.relplot( x= 'Total Conta', y = 'Gorjeta', data=Base_Dados, hue='Sexo');
#insigth = Mulheres dão gorjetas mais altas 
```

```python
# Grafico de linha com 2 eixos
sns.relplot( x = 'Total Conta', y = 'Gorjeta', data=Base_Dados, kind = 'line');
```

```python

sns.relplot( x = 'Total Conta', y = 'Gorjeta', data=Base_Dados, kind = 'line', hue='Fumante');
# aparentemente as pessoas fumantes dão gorjetas mais altas 
```

```python
sns.histplot(data=Base_Dados, x = 'Total Conta')
```

```python
sns.histplot(data=Base_Dados, x = 'Total Conta', hue='Fumante')
```
