```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

* Grafico de Linhas 

```python
np.random.seed(7)

# vetor com números inteiros aleatórios entre 1, 1500, com 10 amostras 

y = np.random.randint(low = 1, high= 1500, size=10)
print(y)
```

```python
# o plot gera um gráfico de linhas aonde i eixo y são dos números aleatórios gerados de 1 até 1500 e o eixo x é o indice da posição dos valores na lista gerada.
plt.plot(y)
```

```python
# insere a primeira linha no plot 

plt.plot(y, color = '#749187', marker = 'o', ms = 5, mec = 'k', markerfacecolor = 'w', ls = '-.')

# insere a segunda linha no plot 
plt.plot(y*2, marker = '+', color = 'k', ms = 5 )

# rótilos
plt.xlabel('Eixo X', color = 'red', size = 12)
plt.ylabel('Eixo Y')
plt.title('Titulo', loc = 'left')

# gridlines

plt.grid(axis = 'y', color = 'gray', linestyle = '--', linewidth = 1, alpha = 0.8)
plt.grid(axis = 'x', color = 'gray', linestyle = '--', linewidth = 1, alpha = 0.8)

```

```python
np.random.seed(6)

x = np.arange(1, 11)
y1 = np.random.randint(1, 400, 10)
y2 = np.random.randint(150, 500, 10)
y3 = np.random.randint(200, 600, 10)

plt.figure(figsize = (15, 5))
plt.suptitle('Figura', fontsize = 15)

plt.subplot(1, 3, 1) # diz que o subplot tem uma linha e 3 colunas e que o subplot vai entrar na primeira coluna
plt.plot(x, y1, color='k')
plt.title('Subplot 1', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.subplot(1, 3, 2) # diz que o subplot tem uma linha e 3 colunas e que o subplot vai entrar na segunda coluna
plt.plot(x, y2, color='r')
plt.title('Subplot 2', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.subplot(1, 3, 3) # diz que o subplot tem uma linha e 3 colunas e que o subplot vai entrar na primeira coluna
plt.plot(x, y3, color='g')
plt.title('Subplot 3', pad = 10)
plt.xlabel('Eixo X')
plt.ylabel('Eixo Y')

plt.tight_layout(pad=1)
plt.show()
```

```python
fig, ax = plt.subplots(1, 3, figsize = (15, 5))
fig.suptitle('Figure')
    
ax[0].plot(x, y1, color='black')
ax[1].plot(x, y2, color='red')
ax[2].plot(x, y3, color='green')

for i in range(3):
    ax[i].set(title = f'Subplot {i + 1}', xlabel = 'Eixo x', ylabel = 'Eixo y')

```
